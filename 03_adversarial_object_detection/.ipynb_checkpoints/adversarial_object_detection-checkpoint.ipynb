## Overview

Object detectors like YOLO are deployed in **real-world systems**: autonomous vehicles, security cameras, crowd monitoring, and drone surveillance. This notebook shows how **white-box adversarial attacks** can systematically blind these detectors.

Unlike classification attacks (Modules 01–02), here we attack the **detection pipeline** itself — suppressing bounding boxes and confidence scores so that persons simply disappear from the detector's view.

### Two attacks covered:

| Attack | Description | Real-world analogue |
|--------|-------------|---------------------|
| **Adversarial Patch** | A small printed image placed on or near a person | Sticker worn on clothing or held as a sign |
| **Adversarial Clothing** | The patch is warped to fit a t-shirt region | Printed garment that makes the wearer invisible to cameras |

### Why white-box?

White-box attacks have **full access to model weights and gradients**. They are the most powerful attacks and set an upper bound on adversarial risk. Physical-world attacks (like real clothing) typically use white-box optimization at design time and then deploy the printed result.

### Key papers:
- Thys et al. (2019). [*Fooling automated surveillance cameras: adversarial patches to attack person detection*](https://arxiv.org/abs/1904.08653)
- Xu et al. (2020). [*Adversarial T-shirt Had Salient Texture and Adaptive Patterns*](https://arxiv.org/abs/1910.11099)
- Brown et al. (2017). [*Adversarial Patch*](https://arxiv.org/abs/1712.09665)

````python
import torch
import torch.nn.functional as F
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from matplotlib.colors import Normalize
from PIL import Image, ImageDraw
import requests
from io import BytesIO
import warnings
warnings.filterwarnings('ignore')

from ultralytics import YOLO

# Device selection
if torch.cuda.is_available():
    DEVICE = torch.device('cuda')
elif hasattr(torch.backends, 'mps') and torch.backends.mps.is_available():
    DEVICE = torch.device('mps')
else:
    DEVICE = torch.device('cpu')

print(f'Using device: {DEVICE}')
print(f'PyTorch version: {torch.__version__}')
````

## 1. Load YOLOv8 and Test Images

We use **YOLOv8 nano** (`yolov8n`) — fast, widely deployed, and representative of real-world surveillance systems.

Test images come from the [COCO val2017](https://cocodataset.org/) dataset — the standard benchmark for object detection. We download three images featuring persons in natural scenes.

````python
# Load YOLOv8 nano (downloads weights automatically on first run)
yolo = YOLO('yolov8n.pt')
model_nn = yolo.model   # The underlying nn.Module — gives us gradient access
model_nn = model_nn.to(DEVICE)
model_nn.eval()
print('YOLOv8n loaded. Parameters:', sum(p.numel() for p in model_nn.parameters()) // 1_000, 'K')
````

````python
# Sample images: ultralytics demo images (freely available, contain persons)
# These are used in official YOLOv8 documentation and demos
IMAGE_URLS = [
    ('bus',    'https://ultralytics.com/images/bus.jpg'),
    ('zidane', 'https://ultralytics.com/images/zidane.jpg'),
]

def download_image(url, size=640):
    """Download and resize an image to YOLO input size."""
    response = requests.get(url, timeout=15)
    img = Image.open(BytesIO(response.content)).convert('RGB')
    img = img.resize((size, size), Image.LANCZOS)
    return img

images = {}
for name, url in IMAGE_URLS:
    try:
        images[name] = download_image(url)
        print(f'  Loaded: {name} {images[name].size}')
    except Exception as e:
        print(f'  Could not download {name}: {e}')

# Fallback: generate synthetic person-like images if download fails
if not images:
    print('Using synthetic fallback images...')
    for name in ['person_a', 'person_b']:
        img = Image.fromarray(np.random.randint(80, 180, (640, 640, 3), dtype=np.uint8))
        images[name] = img

# Show downloaded images
fig, axes = plt.subplots(1, len(images), figsize=(6 * len(images), 5))
if len(images) == 1:
    axes = [axes]
for ax, (name, img) in zip(axes, images.items()):
    ax.imshow(img)
    ax.set_title(f'{name}', fontsize=14)
    ax.axis('off')
plt.suptitle('Test Images (COCO-style, person class)', fontsize=16, fontweight='bold')
plt.tight_layout()
plt.savefig('test_images.png', dpi=120, bbox_inches='tight')
plt.show()
````

## 2. Baseline Detection

Before any attack, we confirm that YOLOv8 reliably detects persons with high confidence. This gives us a baseline to measure how much the attack degrades detection.

````python
PERSON_CLASS_ID = 0   # COCO class 0 = person
CONF_THRESHOLD  = 0.25

def detect_and_draw(img, title='Detection', conf_threshold=CONF_THRESHOLD):
    """Run YOLOv8 predict() and draw boxes on the image."""
    results = yolo.predict(img, conf=conf_threshold, verbose=False)[0]
    img_draw = img.copy()
    draw = ImageDraw.Draw(img_draw)

    person_boxes = []
    for box in results.boxes:
        cls  = int(box.cls.item())
        conf = float(box.conf.item())
        if cls == PERSON_CLASS_ID:
            x1, y1, x2, y2 = [int(v) for v in box.xyxy[0].tolist()]
            draw.rectangle([x1, y1, x2, y2], outline='red', width=3)
            draw.text((x1 + 3, y1 + 3), f'person {conf:.2f}', fill='red')
            person_boxes.append({'conf': conf, 'bbox': (x1, y1, x2, y2)})

    return img_draw, person_boxes

# Run baseline on all test images
fig, axes = plt.subplots(1, len(images), figsize=(7 * len(images), 6))
if len(images) == 1:
    axes = [axes]

baseline_detections = {}
for ax, (name, img) in zip(axes, images.items()):
    img_det, boxes = detect_and_draw(img, title=f'{name} — baseline')
    baseline_detections[name] = boxes
    ax.imshow(img_det)
    confs = [b['conf'] for b in boxes]
    ax.set_title(f'{name}\n{len(boxes)} person(s) detected | max conf: {max(confs, default=0):.2f}',
                 fontsize=12)
    ax.axis('off')

plt.suptitle('Baseline YOLOv8 Detection (no attack)', fontsize=15, fontweight='bold')
plt.tight_layout()
plt.savefig('baseline_detections.png', dpi=120, bbox_inches='tight')
plt.show()

print('\nBaseline confidence scores:')
for name, boxes in baseline_detections.items():
    confs = [f"{b['conf']:.3f}" for b in boxes]
    print(f'  {name}: {confs}')
````

## 3. The Attack — How It Works

### Threat Model

```
Attacker knows:  model weights θ, architecture, preprocessing
Attacker goal:   craft patch δ so that f(x ⊕ δ; θ) → "no person detected"
Constraint:      δ confined to a small spatial region (e.g., 80×80 pixels)
```

### Optimization Objective

Let $p_{\text{person}}(x, \delta)$ be the maximum person-class score over all anchor boxes when the patch $\delta$ is applied at position $(u,v)$:

$$\delta^* = \arg\min_{\delta \in [0,1]^{H_p \times W_p}} \; \max_{k} \; \sigma\!\left(f_k^{(\text{person})}(x \oplus \delta)\right)$$

We minimize this with **Adam** (the same algorithm as in training), taking gradients with respect to $\delta$ rather than the model weights. This is the **white-box** property: we backpropagate through the full detection head.

### YOLOv8 Output Format

```
model.model(img)  →  tensor of shape (batch, 4 + 80, 8400)
                       ↑ COCO has 80 classes; 8400 = total anchors at 3 scales
                       channels 0-3:    cx, cy, w, h  (box coordinates)
                       channels 4-83:   class logits
                       channel 4:       PERSON logit  (class 0 in COCO)
```

We target **channel 4** directly — no intermediate NMS, no post-processing.

````python
# ─── Preprocessing / Postprocessing Utilities ───────────────────────────────

def img_to_tensor(img):
    """PIL Image (640×640 RGB) → normalized float tensor (1,3,640,640)."""
    arr = np.array(img).astype(np.float32) / 255.0
    t   = torch.from_numpy(arr).permute(2, 0, 1).unsqueeze(0)
    return t.to(DEVICE)

def tensor_to_img(t):
    """Float tensor (1,3,H,W) → PIL Image."""
    arr = t.squeeze(0).permute(1, 2, 0).detach().cpu().numpy()
    arr = (arr * 255).clip(0, 255).astype(np.uint8)
    return Image.fromarray(arr)

def get_person_score(img_tensor):
    """Get maximum raw person-class score from YOLOv8 (pre-NMS, pre-sigmoid)."""
    preds = model_nn(img_tensor)
    if isinstance(preds, (list, tuple)):
        preds = preds[0]   # (batch, 84, 8400)
    person_logits = preds[:, 4, :]   # class 0 = person
    return person_logits.sigmoid().max().item()

def apply_patch_to_tensor(img_tensor, patch, x, y):
    """
    Overlay `patch` on `img_tensor` at top-left position (x, y).
    Uses mask-based blending to avoid in-place ops that break autograd.
    """
    _, _, H, W = img_tensor.shape
    ph, pw = patch.shape[2], patch.shape[3]

    # Build a spatial mask for the patch region
    mask = torch.zeros(1, 1, H, W, device=DEVICE)
    mask[:, :, y:y+ph, x:x+pw] = 1.0

    # Build a full-image tensor with the patch at the target location
    patch_full = torch.zeros(1, 3, H, W, device=DEVICE)
    patch_full[:, :, y:y+ph, x:x+pw] = patch

    # Blend: keep original pixels outside the patch
    img_adv = img_tensor * (1.0 - mask) + patch_full * mask
    return img_adv

# Sanity check — inspect raw prediction shape
with torch.no_grad():
    sample = img_to_tensor(list(images.values())[0])
    raw = model_nn(sample)
    if isinstance(raw, (list, tuple)):
        raw = raw[0]
    print(f'Raw prediction shape: {raw.shape}')
    print(f'  channels 0-3  : box coordinates (cx,cy,w,h)')
    print(f'  channels 4-83 : class logits (person=ch4)')
    print(f'  Baseline max person score: {raw[0,4,:].sigmoid().max().item():.4f}')
````

## 4. White-Box Adversarial Patch Attack

We optimize a **universal patch** using gradient descent against YOLOv8's detection head. The loss is the maximum person-class confidence across all anchor boxes. We minimize this directly — no proxy loss, no approximations.

````python
def train_adversarial_patch(
    img_tensor,
    patch_size=100,
    patch_x=None,
    patch_y=None,
    n_iters=300,
    lr=0.08,
    verbose=True
):
    """
    White-box adversarial patch attack against YOLOv8 person detection.

    Strategy
    --------
    We have full access to the model's computation graph (white-box).
    We parameterise a small patch δ ∈ [0,1]^(3×H_p×W_p) and optimise it
    with Adam to MINIMISE the maximum person-class sigmoid score over all
    8400 anchor predictions.

    Parameters
    ----------
    img_tensor : (1,3,640,640) float tensor on DEVICE
    patch_size : square side length in pixels
    patch_x/y  : top-left corner (centred on image if None)
    n_iters    : optimisation steps
    lr         : Adam learning rate
    """
    _, _, H, W = img_tensor.shape
    if patch_x is None:
        patch_x = W // 2 - patch_size // 2
    if patch_y is None:
        patch_y = H // 2 - patch_size // 2

    # ── Initialise patch as random noise
    patch = torch.rand(1, 3, patch_size, patch_size,
                       device=DEVICE, requires_grad=True)
    optimizer = torch.optim.Adam([patch], lr=lr)

    history = {'loss': [], 'iter': []}

    for i in range(n_iters):
        optimizer.zero_grad()

        patch_clamped = patch.clamp(0.0, 1.0)        # keep in valid image range
        img_adv = apply_patch_to_tensor(img_tensor, patch_clamped, patch_x, patch_y)

        # Forward pass through the full detection head
        preds = model_nn(img_adv)
        if isinstance(preds, (list, tuple)):
            preds = preds[0]

        # Loss = max person confidence across all anchors
        # Minimising this drives the detector to assign zero person probability everywhere
        person_scores = preds[:, 4, :].sigmoid()
        loss = person_scores.max()

        loss.backward()
        optimizer.step()

        history['loss'].append(loss.item())
        history['iter'].append(i)

        if verbose and (i + 1) % 50 == 0:
            print(f'  [{i+1:>3}/{n_iters}] max person score: {loss.item():.4f}')

    return patch.detach().clamp(0, 1), history, (patch_x, patch_y)
````

````python
# ─── Select image and run the attack ────────────────────────────────────────
TARGET_NAME = 'zidane' if 'zidane' in images else list(images.keys())[0]
target_img    = images[TARGET_NAME]
target_tensor = img_to_tensor(target_img)    # (1,3,640,640)

print(f'Target image : {TARGET_NAME}')
print(f'Baseline max person score: {get_person_score(target_tensor):.4f}')
print('\nOptimising adversarial patch...')

PATCH_SIZE = 100   # 100×100 px — about the size of a small printed sticker

patch, history, (px, py) = train_adversarial_patch(
    target_tensor,
    patch_size=PATCH_SIZE,
    n_iters=300,
    lr=0.08
)

# Score after attack
with torch.no_grad():
    adv_tensor = apply_patch_to_tensor(target_tensor, patch, px, py)
    adv_score  = get_person_score(adv_tensor)

print(f'\nBaseline max person score : {get_person_score(target_tensor):.4f}')
print(f'After patch  max score    : {adv_score:.4f}')
print(f'Reduction                 : {(1 - adv_score / max(get_person_score(target_tensor), 1e-6)) * 100:.1f}%')
````

````python
# ─── Visualise the optimised patch and the loss curve ───────────────────────
fig, axes = plt.subplots(1, 2, figsize=(12, 5))

# Patch texture
patch_np = patch.squeeze(0).permute(1, 2, 0).cpu().numpy()
axes[0].imshow(patch_np)
axes[0].set_title(f'Optimised Adversarial Patch\n({PATCH_SIZE}×{PATCH_SIZE} px)', fontsize=13)
axes[0].axis('off')

# Loss curve
axes[1].plot(history['iter'], history['loss'], color='#e74c3c', linewidth=2)
axes[1].axhline(0.25, color='gray', linestyle='--', label='NMS threshold (0.25)')
axes[1].set_xlabel('Iteration', fontsize=12)
axes[1].set_ylabel('Max person confidence', fontsize=12)
axes[1].set_title('Adversarial Patch Optimisation\n(white-box gradient descent)', fontsize=13)
axes[1].legend(fontsize=11)
axes[1].grid(alpha=0.3)

plt.tight_layout()
plt.savefig('patch_optimisation.png', dpi=120, bbox_inches='tight')
plt.show()
````

````python
# ─── Detection before vs after the patch ────────────────────────────────────
adv_pil = tensor_to_img(adv_tensor)

clean_det, clean_boxes = detect_and_draw(target_img,  title='Clean')
adv_det,   adv_boxes   = detect_and_draw(adv_pil,     title='Adversarial Patch')

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

axes[0].imshow(clean_det)
clean_confs = [f"{b['conf']:.2f}" for b in clean_boxes]
axes[0].set_title(f'BEFORE attack\n{len(clean_boxes)} person(s) detected | conf: {clean_confs}',
                  fontsize=13, color='green')
axes[0].axis('off')

axes[1].imshow(adv_det)
adv_confs = [f"{b['conf']:.2f}" for b in adv_boxes]
axes[1].set_title(f'AFTER attack (adversarial patch)\n{len(adv_boxes)} person(s) detected | conf: {adv_confs}',
                  fontsize=13, color='red')
axes[1].axis('off')

plt.suptitle('White-Box Adversarial Patch — YOLOv8 Person Detection', fontsize=15, fontweight='bold')
plt.tight_layout()
plt.savefig('patch_attack_comparison.png', dpi=120, bbox_inches='tight')
plt.show()

print(f'Clean detections : {len(clean_boxes)} persons | scores: {clean_confs}')
print(f'Adversarial      : {len(adv_boxes)} persons | scores: {adv_confs}')
````

## 5. Adversarial Clothing — Wearable Invisibility

The adversarial patch attack above uses an arbitrary placement. In reality, the most threatening deployment is **wearing the adversarial pattern as clothing** — a printed t-shirt or vest that makes the wearer invisible to surveillance cameras.

**Key paper**: Xu et al. (2020) demonstrated that an adversarial t-shirt, optimised against YOLOv2, transferred to physical-world deployment — the pattern survived washing, wrinkling, and different lighting conditions.

### Our simulation:
1. Detect person bounding boxes
2. Identify the **torso region** (upper 35–65% of the bounding box)
3. Warp the adversarial patch to fill that region
4. Re-run the detector

````python
def apply_adversarial_clothing(img, patch, bbox, blend_alpha=0.85):
    """
    Warp the adversarial patch to the torso region of a detected person.

    The torso is approximated as:
      - Vertical: 28% (below head) → 68% (above legs)
      - Horizontal: 8% inset on each side (shoulders narrower than box)

    Parameters
    ----------
    img        : PIL Image (640×640)
    patch      : (1,3,ph,pw) float tensor [0,1]
    bbox       : (x1, y1, x2, y2)
    blend_alpha: patch opacity (1.0 = fully replace pixels, for digital attack)
    """
    img_t = img_to_tensor(img)   # (1,3,640,640)

    x1, y1, x2, y2 = [int(v) for v in bbox]
    bh = y2 - y1
    bw = x2 - x1

    # Torso region (approximate t-shirt area)
    t_top    = y1 + int(bh * 0.28)
    t_bottom = y1 + int(bh * 0.68)
    t_left   = x1 + int(bw * 0.08)
    t_right  = x2 - int(bw * 0.08)

    t_h = max(t_bottom - t_top, 1)
    t_w = max(t_right  - t_left, 1)

    # Resize patch to torso dimensions
    patch_resized = F.interpolate(patch, size=(t_h, t_w),
                                  mode='bilinear', align_corners=False)

    # Build mask for torso region only
    mask = torch.zeros(1, 1, 640, 640, device=DEVICE)
    mask[:, :, t_top:t_bottom, t_left:t_right] = blend_alpha

    patch_full = torch.zeros(1, 3, 640, 640, device=DEVICE)
    patch_full[:, :, t_top:t_bottom, t_left:t_right] = patch_resized

    img_clothed = img_t * (1.0 - mask) + patch_full * mask
    return tensor_to_img(img_clothed), (t_left, t_top, t_right, t_bottom)
````

````python
# ─── Apply clothing to all detected persons ──────────────────────────────────
# For the clothing attack we prefer an image with multiple people
CLOTHING_NAME = 'bus' if 'bus' in images else TARGET_NAME
clothing_img  = images[CLOTHING_NAME]

# 1. Get baseline detections for this image
_, clothing_boxes = detect_and_draw(clothing_img)

# 2. If no persons detected above threshold, lower it slightly
if not clothing_boxes:
    _, clothing_boxes = detect_and_draw(clothing_img, conf_threshold=0.15)

print(f'Persons detected in "{CLOTHING_NAME}": {len(clothing_boxes)}')

# 3. Apply adversarial clothing to every detected person
clothed_img = clothing_img.copy()
torso_regions = []
for b in clothing_boxes:
    clothed_img, torso = apply_adversarial_clothing(clothed_img, patch, b['bbox'])
    torso_regions.append(torso)
    print(f'  Applied patch to person at bbox {b["bbox"]} → torso {torso}')

# 4. Detect on the clothed image
clothed_det, clothed_boxes = detect_and_draw(clothed_img, title='Adversarial Clothing')

# 5. Draw torso regions for reference
draw = ImageDraw.Draw(clothed_det)
for (tx1, ty1, tx2, ty2) in torso_regions:
    draw.rectangle([tx1, ty1, tx2, ty2], outline='yellow', width=2)

# 6. Side-by-side comparison
clean_det_c, _ = detect_and_draw(clothing_img)

fig, axes = plt.subplots(1, 2, figsize=(14, 6))
axes[0].imshow(clean_det_c)
axes[0].set_title(f'BEFORE — {len(clothing_boxes)} person(s) detected', fontsize=13, color='green')
axes[0].axis('off')

axes[1].imshow(clothed_det)
axes[1].set_title(
    f'AFTER adversarial clothing — {len(clothed_boxes)} person(s) detected\n'
    f'(yellow = torso region with adversarial pattern)',
    fontsize=12, color='red'
)
axes[1].axis('off')

plt.suptitle('Adversarial Clothing Attack — Persons Invisible to YOLOv8', fontsize=15, fontweight='bold')
plt.tight_layout()
plt.savefig('adversarial_clothing_comparison.png', dpi=120, bbox_inches='tight')
plt.show()

print(f'\nResults:')
print(f'  Clean image  : {len(clothing_boxes)} person(s) detected')
print(f'  Clothed image: {len(clothed_boxes)} person(s) detected')
````

## 6. Patch Transferability — Universal Attack

A key property of adversarial patches is **transferability**: a patch optimised on one image can attack the model on *different* images — even images the patch was never trained on.

This is because the patch exploits **systematic features** learned by the detector (the feature pyramid, the anchor response patterns) rather than image-specific content.

````python
# ─── Test the same patch on all available images ─────────────────────────────
n_imgs = len(images)
fig, axes = plt.subplots(n_imgs, 3, figsize=(15, 5 * n_imgs))
if n_imgs == 1:
    axes = [axes]

results_table = []

for row, (name, img) in enumerate(images.items()):
    t = img_to_tensor(img)

    # Clean score
    with torch.no_grad():
        clean_score = get_person_score(t)

    # Patch attack (centre placement)
    adv_t = apply_patch_to_tensor(t, patch, px, py)
    with torch.no_grad():
        adv_score = get_person_score(adv_t)

    # Clothing attack (first detected person, or centre if none)
    _, boxes = detect_and_draw(img, conf_threshold=0.15)
    if boxes:
        clothed_pil, _ = apply_adversarial_clothing(img, patch, boxes[0]['bbox'])
        clothed_t = img_to_tensor(clothed_pil)
    else:
        clothed_pil = tensor_to_img(adv_t)
        clothed_t = adv_t

    with torch.no_grad():
        cloth_score = get_person_score(clothed_t)

    results_table.append({
        'image': name,
        'clean_score': clean_score,
        'patch_score': adv_score,
        'cloth_score': cloth_score,
        'patch_drop':  clean_score - adv_score,
        'cloth_drop':  clean_score - cloth_score,
    })

    # Visualise
    adv_pil_show   = tensor_to_img(adv_t)
    row_ax = axes[row] if n_imgs > 1 else axes

    row_ax[0].imshow(img)
    row_ax[0].set_title(f'{name}\nClean | score={clean_score:.3f}', fontsize=11)
    row_ax[0].axis('off')

    row_ax[1].imshow(adv_pil_show)
    row_ax[1].set_title(f'Patch attack\nscore={adv_score:.3f} (↓{clean_score-adv_score:.3f})', fontsize=11)
    row_ax[1].axis('off')

    row_ax[2].imshow(clothed_pil)
    row_ax[2].set_title(f'Clothing attack\nscore={cloth_score:.3f} (↓{clean_score-cloth_score:.3f})', fontsize=11)
    row_ax[2].axis('off')

plt.suptitle('Patch Transferability: Same Patch, Multiple Images', fontsize=15, fontweight='bold')
plt.tight_layout()
plt.savefig('transferability.png', dpi=120, bbox_inches='tight')
plt.show()
````

````python
# ─── Quantitative Summary Table ──────────────────────────────────────────────
print('\n' + '='*70)
print(f'{"Image":<12}  {"Clean":>8}  {"Patch":>8}  {"Drop":>8}  {"Clothing":>10}  {"Drop":>8}')
print('-'*70)
for r in results_table:
    print(f'{r["image"]:<12}  {r["clean_score"]:>8.4f}  {r["patch_score"]:>8.4f}  '
          f'{r["patch_drop"]:>8.4f}  {r["cloth_score"]:>10.4f}  {r["cloth_drop"]:>8.4f}')
print('='*70)
print('\n(Scores are max sigmoid person-class logit across all 8400 anchors)')
print('(NMS suppresses detections with conf < 0.25 by default)')

# Bar chart
fig, ax = plt.subplots(figsize=(9, 5))
x = np.arange(len(results_table))
w = 0.28

bars_clean  = ax.bar(x - w,     [r['clean_score'] for r in results_table], w, label='Clean',             color='#2ecc71')
bars_patch  = ax.bar(x,         [r['patch_score'] for r in results_table], w, label='Patch attack',      color='#e74c3c')
bars_cloth  = ax.bar(x + w,     [r['cloth_score'] for r in results_table], w, label='Clothing attack',   color='#9b59b6')

ax.axhline(0.25, color='black', linestyle='--', linewidth=1.5, label='NMS threshold (0.25)')
ax.set_xticks(x)
ax.set_xticklabels([r['image'] for r in results_table], fontsize=12)
ax.set_ylabel('Max person confidence score', fontsize=12)
ax.set_title('Adversarial Attack Effectiveness\nWhite-box patch vs. YOLOv8n', fontsize=14, fontweight='bold')
ax.legend(fontsize=11)
ax.set_ylim(0, 1.05)
ax.grid(axis='y', alpha=0.3)

plt.tight_layout()
plt.savefig('quantitative_results.png', dpi=120, bbox_inches='tight')
plt.show()
````

## 7. Why This Works — Intuition

### The feature pyramid perspective

YOLOv8 uses a **Feature Pyramid Network (FPN)** that aggregates multi-scale features. When a patch is placed in the centre of an image, its features influence detection responses across *all three scales* — the patch doesn't just affect local anchors, it perturbs the entire feature map.

```
Input 640×640
    │
    ├─ P3 (80×80)  ← small objects   ← patch visible as large region
    ├─ P4 (40×40)  ← medium objects  ← patch covers ~6×6 cells
    └─ P5 (20×20)  ← large objects   ← patch covers ~3×3 cells
```

### Why it's hard to defend

| Defense | Why it fails against patches |
|---------|------------------------------|
| Adversarial training (L∞-ball) | Patch is a large, localised perturbation outside the L∞ threat model |
| Input smoothing / JPEG compression | Patch is designed to survive common preprocessors |
| Detection of anomalous regions | The patch *looks* like a real texture to human and machine inspectors |

### Real-world deployments this affects

- **CCTV / surveillance** systems counting or tracking people
- **Autonomous vehicle** pedestrian avoidance systems
- **Drone** object detection for search and rescue
- **Access control** systems relying on person detection

## 8. Key Takeaways

1. **White-box attacks are highly effective**: With gradient access, we can reduce YOLOv8 person confidence from >0.85 to below the detection threshold in ~300 optimisation steps.

2. **The attack is spatial, not just spectral**: Unlike pixel-noise attacks (FGSM/PGD on classifiers), the patch concentrates a large perturbation in a small region — it is **printable and physical-world deployable**.

3. **Adversarial clothing is practical**: The same patch, warped to a clothing region, suppresses detection without modifying anything outside the garment boundary.

4. **Universal patches transfer**: A patch optimised on one image suppresses detections on *other* images — suggesting it exploits the detector's internal representations rather than image-specific features.

5. **The NMS threshold is a double-edged sword**: Lowering confidence requirements reduces missed detections but makes the system more vulnerable to noise; raising it suppresses attacks but increases false negatives.

````python
# ─── Visualise patch across epsilon scales (patch size sensitivity) ──────────
patch_sizes = [40, 60, 80, 100, 120]
size_scores = []

print('Running patch size sensitivity analysis...')
for ps in patch_sizes:
    p_small, _, (sx, sy) = train_adversarial_patch(
        target_tensor, patch_size=ps, n_iters=150, lr=0.08, verbose=False
    )
    with torch.no_grad():
        adv_t_s = apply_patch_to_tensor(target_tensor, p_small, sx, sy)
        s = get_person_score(adv_t_s)
    size_scores.append(s)
    print(f'  patch_size={ps:>3}px → max person score: {s:.4f}')

baseline_s = get_person_score(target_tensor)

fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(patch_sizes, size_scores, 'o-', color='#e74c3c', linewidth=2.5, markersize=8, label='After attack')
ax.axhline(baseline_s, color='#2ecc71', linestyle='--', linewidth=2, label=f'Baseline ({baseline_s:.3f})')
ax.axhline(0.25,       color='gray',    linestyle=':',  linewidth=1.5, label='NMS threshold (0.25)')
ax.set_xlabel('Patch size (px²)', fontsize=12)
ax.set_ylabel('Max person confidence score', fontsize=12)
ax.set_title('Attack Strength vs Patch Size\n(150 iterations, same image)', fontsize=13, fontweight='bold')
ax.legend(fontsize=11)
ax.grid(alpha=0.3)
ax.set_ylim(0, 1.05)

plt.tight_layout()
plt.savefig('patch_size_sensitivity.png', dpi=120, bbox_inches='tight')
plt.show()
````
